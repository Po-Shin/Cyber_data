### Compute the connections( Dst_IP) of the infected host (147.32.84.165)

```python
df = pd.read_pickle('capture20110815-2.pcap.netflow_ps.pickle')
df['Dst_IP'] = df['Dst_IP:Port'].str.split(':',expand=True)[0]
df['Src_IP'] = df['Src_IP:Port'].str.split(':',expand=True)[0]
df = df[df['Src_IP']=='147.32.84.165'].reset_index(drop=True)
print(len(df))
# df['Dst_IP'].value_counts(normalize = True).head(10)
df['Dst_IP'].value_counts().head(10)
```



## Min_wise sampling

```python
#uniform_sampling
reservoir_size = 0
for reservoir_size in np.arange(100,1100,300):
    Freq = df['Dst_IP'].value_counts(normalize = True)
    Freq = np.ceil(Freq*reservoir_size)
#     print(Freq)
    IP_list = list(df['Dst_IP'].value_counts(normalize = True).index)
    reservoir = []
    
    #########compute the elapsed time --start
    time_start = time.clock()
    
    ###create reservoir
    count = 0
    for i in range(len(IP_list)):
        if count < reservoir_size:
            for k in range(np.int(Freq[i])):
                reservoir.append(IP_list[i])
            count += Freq[i]
#     print(reservoir)


    ######## Do min-wise sampling
    m = np.random.uniform(0,1,len(df)+reservoir_size)
    for i in range(len(df)):
        reservoir.append(df.loc[i,'Dst_IP'])
        del reservoir[np.argmax(m[0:reservoir_size+1])]
        m = np.delete(m,np.argmax(m[0:reservoir_size+1]))
        
    ######## sort the frequency
    counter = collections.Counter(reservoir)
    print("time_elapsed %.3f"%(time.clock() - time_start))
    print("size = %d"%reservoir_size)
    print(collections.OrderedDict(counter.most_common(10)),'\n')
#     print((list(collections.OrderedDict(counter.most_common()).values())/reservoir_size)[0:10])
```




## Count_min_sketch

```python
######init
count_min_w = 1000
count_min_d = 3
count_min_table =np.zeros((count_min_d,count_min_w))
# print(count_min_table.shape)

IP_hash = {}
time_start = time.clock()

###### Count-min sketch
for i in (df['Dst_IP']):
    #1:md5
    md5 = hashlib.md5(str(i).encode())
    index = (int(md5.hexdigest(),16)%count_min_w)
    count_min_table[0,index] +=1
    if str(str(i)+'_md5') not in IP_hash.keys():
        IP_hash[str(i)+'_md5'] = [0,index]
    #sha1
    sha1 = hashlib.sha1(str(i).encode())
    index = (int(sha1.hexdigest(),16)%count_min_w)
    count_min_table[1,index] +=1 
    if str(str(i)+'_sha1') not in IP_hash.keys():
        IP_hash[str(i)+'_sha1'] = [1,index]
    #sha224
    sha224 = hashlib.sha224(str(i).encode())
    index = (int(sha224.hexdigest(),16)%count_min_w)
    count_min_table[2,index] +=1
    if str(str(i)+'_sha224') not in IP_hash.keys():
        IP_hash[str(i)+'_sha224'] = [2,index]
print("done")

freq_IPs = []

######find frequency:
IP_count = 0
for i in IP_list:
    i_md5 = count_min_table[IP_hash[i+'_md5'][0],IP_hash[i+'_md5'][1]]
    i_sha1 = count_min_table[IP_hash[i+'_sha1'][0],IP_hash[i+'_sha1'][1]]
    i_sha224 = count_min_table[IP_hash[i+'_sha224'][0],IP_hash[i+'_sha224'][1]]
    if IP_count < 10:
        print(i,' : ',np.min([i_md5,i_sha1,i_sha224]))
        IP_count+=1

print("time_elapsed %.3f"%(time.clock() - time_start))
```
